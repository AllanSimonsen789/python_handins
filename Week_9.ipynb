## Exercise 1


### Task 1 & 2
1) Create a python notebook, and download the dataset from the link above.

2) Use python to create a new file and copy all the data from the .txt.gz into a .txt file.

```python
import gzip
import shutil

with gzip.open("./week9modules/Wiki-Vote.txt.gz", "rb") as gz_object:
    with open("./week9modules/Wiki-Vote.txt", "wb") as txt_object:
        shutil.copyfileobj(gz_object, txt_object)
```

### Task 3
3) Use networkx to add the nodes from the file we just created: NetworkX_Documentation

```python
import networkx as nx

graph = nx.read_edgelist("./week9modules/Wiki-Vote.txt")

print(nx.info(graph))
```

### Task 4
4) Make a random number generator that is able to create a number between 1 and the length of your Undirected graphs.

```python
import random

idchosen = random.randint(1, len(graph))

print(idchosen)
```

### Task 5
5) Print all the neighbouring nodes, of the randomly generated id.

```python
print(graph[str(idchosen)])
```

### Task 6
6) Make a visual presentation of the randomly generated id in the notebooks. nx.draw()

```python
plotting_data = graph.subgraph(graph[str(idchosen)])
nx.draw_shell(plotting_data, with_labels=True)
```
